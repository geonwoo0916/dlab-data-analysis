```python
# 키몸무게 데이터 읽어오기
import pandas as pd
import matplotlib.pyplot as plt

df_k = pd.read_csv('/content/키_몸무게_2019.csv')
df_k
```

데이터 행 열 바꾸기

```python
df_k = df_k.set_index('Unnamed: 0')
df_k
```

```python
df_k = df_k.T
df_k
```

```python
df_k.columns = ['신장', '체중', '허리둘레']
df_k
```

```python
df_k['신장'].unique()
```

```python
df_k[ df_k['신장'] == 160 ]
df_160 = df_k[ df_k['신장'] == 160 ]
df_160
```

```python
df_160['체중'].sum() / len(df_160)
```

```python
df_k.groupby('신장')['체중'].count()
```

```python
df_k.groupby('체중')['허리둘레'].mean()
```

```python
import seaborn as sns

sns.boxplot(df_k)
plt.show()
```

```python
sns.boxplot(data=df_k, x = '신장', y = '허리둘레')
plt.show()
```

```python
# apply 활용

def 체중구분(체중):
  if 체중 < 55:
    return '마름'
  elif 체중 < 70:
    return '보통'
  else:
    return '통통'

# 열 추가법
df_k['구분'] = df_k['체중'].apply(체중구분)
df_k
```

```python
sns.scatterplot(data = df_k, x='신장', y='체중', color='red')
plt.show()
# plt.scatterplot(x = df_k['신장'], y = df_k['체중'], color='red')
# plt.show()
```

```python
# 맥도날드 세트메뉴 data 가져오기
df_m = pd.read_csv('/content/맥도날드 세트메뉴 칼로리.csv')
df_m
```
